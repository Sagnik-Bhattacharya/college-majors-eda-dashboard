```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style('whitegrid')
```

```python
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:,.2f}".format)
```

```python
all_ages = pd.read_csv("../data/raw/all-ages.csv")
recent_grads = pd.read_csv("../data/raw/recent-grads.csv")
grad_students = pd.read_csv("../data/raw/grad-students.csv")
women_stem = pd.read_csv("../data/raw/women-stem.csv")
majors_list = pd.read_csv("../data/raw/majors-list.csv")
```

```python
datasets = {
    "all_ages": all_ages,
    "recent_grads": recent_grads,
    "grad_students": grad_students,
    "women_stem": women_stem,
    "majors_list": majors_list
}

for name, df in datasets.items():
    print(f"\n{name.upper()}")
    display(df.head())
    print(df.info())
```

```python
for df in datasets.values():
    if "index" in df.columns:
        df.drop(columns="index", inplace=True)
```

```python
assert all_ages["Major_code"].is_unique
assert recent_grads["Major_code"].is_unique
assert grad_students["Major_code"].is_unique
assert women_stem["Major_code"].is_unique
```

- `Each dataset maintains a one-to-one relationship with Major_code, enabling safe joins later in SQL.`

```python
for name, df in datasets.items():
    print(f"\n{name}")
    print(df.isna().sum())
```

```python
salary_cols = ["Median", "P25th", "P75th"]

for col in salary_cols:
    if col in all_ages.columns:
        all_ages[col] = all_ages[col].astype("Int64")
```

```python
recent_grads["Median"] = recent_grads["Median"].astype("Int64")
grad_students["Grad_median"] = grad_students["Grad_median"].astype("Int64")
grad_students["Nongrad_median"] = grad_students["Nongrad_median"].astype("Int64")
```

```python
all_ages["employment_rate"] = (
    all_ages["Employed"] / all_ages["Total"]
)
```

```python
all_ages["full_time_share"] = (
    all_ages["Employed_full_time_year_round"] / all_ages["Employed"]
)
```

```python
all_ages["salary_spread"] = (
    all_ages["P75th"] - all_ages["P25th"]
)
```

```python
grad_students["salary_premium"] = (
    grad_students["Grad_median"] - grad_students["Nongrad_median"]
)
```

```python
all_ages["Major_category"] = all_ages["Major_category"].str.strip()
recent_grads["Major_category"] = recent_grads["Major_category"].str.strip()
```
